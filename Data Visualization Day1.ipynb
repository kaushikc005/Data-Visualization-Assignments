```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.plot([1,2,3],[2,3,4])
plt.show()
```

```python
x=[1,2,3]
y=[2,3,4]

plt.title("Simple")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,label="2x",linestyle="--",color="b",linewidth=1,marker="*",markersize=10,markeredgecolor="r")
plt.savefig("abc.png",dpi=300)
plt.legend()
plt.show()
```

```python
## Assignment 1
x=[1,2,3,4,5]
y=[0,2,4,6,8]

plt.figure(figsize=(6,4),dpi=100)
plt.title("Second Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x,y,label="2x",linestyle=":",color="r",marker="*",markeredgecolor="b")

plt.savefig("ass1.png",dpi=300)
plt.legend()
plt.show()
```

```python
x2=np.arange(0,4,0.5)
print(x2)

x2**2
```

```python
labels=['A','B','C']
values=[1,4,7]

plt.bar(labels,values)
plt.show()
```

```python
labels=['A','B','C']
values=[1,4,7]

plt.figure(figsize=(6,4))

bars=plt.bar(labels,values)

bars[0].set_hatch("o")
bars[1].set_hatch("/")
bars[2].set_hatch("+")

plt.show()
```
